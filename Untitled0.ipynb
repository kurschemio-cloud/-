```python
from IPython.display import HTML

html_content = '''
<!DOCTYPE html>
<html lang="ru">
<head>
<meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>🧪 Продукты реакций простых веществ</title>
    <style>
        :root {
            --primary: #6c5ce7;
            --success: #00b894;
            --error: #d63031;
            --bg: #0a0a1a;
            --card: #1a1a2e;
            --card2: #16213e;
            --accent: #e94560;
            --text: #eee;
            --dim: #888;
        }

        * { margin: 0; padding: 0; box-sizing: border-box; }

        body {
            font-family: 'Segoe UI', system-ui, -apple-system, sans-serif;
            background: var(--bg);
            color: var(--text);
            min-height: 100vh;
            display: flex;
            justify-content: center;
            align-items: center;
            overflow-x: hidden;
        }

        .game-wrapper {
            width: 95%;
            max-width: 700px;
            padding: 20px;
        }

        .header {
            text-align: center;
            margin-bottom: 20px;
        }

        .header h1 {
            font-size: 1.5rem;
            background: linear-gradient(135deg, var(--accent), var(--primary));
            -webkit-background-clip: text;
            -webkit-text-fill-color: transparent;
            margin-bottom: 8px;
        }

        .stats-bar {
            display: flex;
            justify-content: space-between;
            gap: 10px;
            margin-bottom: 20px;
            flex-wrap: wrap;
        }

        .stat-box {
            background: var(--card);
            border: 1px solid #333;
            border-radius: 12px;
            padding: 10px 16px;
            flex: 1;
            min-width: 80px;
            text-align: center;
        }

        .stat-box .label {
            font-size: 0.7rem;
            color: var(--dim);
            text-transform: uppercase;
            letter-spacing: 1px;
        }

        .stat-box .value {
            font-size: 1.5rem;
            font-weight: bold;
            color: var(--accent);
        }

        .stat-box .value.green { color: var(--success); }

        .reaction-card {
            background: linear-gradient(145deg, var(--card), var(--card2));
            border: 1px solid #333;
            border-radius: 20px;
            padding: 30px 20px;
            text-align: center;
            margin-bottom: 20px;
            position: relative;
            overflow: hidden;
        }

        .reaction-card::before {
            content: '';
            position: absolute;
            top: 0; left: 0; right: 0;
            height: 3px;
            background: linear-gradient(90deg, var(--accent), var(--primary), var(--success));
        }

        .reaction-display {
            font-size: 2rem;
            margin: 15px 0;
            display: flex;
            align-items: center;
            justify-content: center;
            gap: 12px;
            flex-wrap: wrap;
        }

        .element-box {
            background: rgba(255,255,255,0.08);
            border: 2px solid var(--primary);
            border-radius: 12px;
            padding: 12px 20px;
            font-weight: bold;
            font-size: 1.8rem;
            min-width: 80px;
        }

        .plus-sign { color: var(--dim); font-size: 1.5rem; }
        .arrow { color: var(--accent); font-size: 1.8rem; }

        .question-mark {
            background: rgba(233, 69, 96, 0.2);
            border: 2px dashed var(--accent);
            border-radius: 12px;
            padding: 12px 25px;
            font-size: 2rem;
            color: var(--accent);
            animation: pulse 1.5s infinite;
        }

        @keyframes pulse {
            0%, 100% { opacity: 1; transform: scale(1); }
            50% { opacity: 0.7; transform: scale(1.05); }
        }

        .conditions-hint {
            font-size: 0.85rem;
            color: var(--dim);
            margin-top: 10px;
            font-style: italic;
            min-height: 1.2em;
        }

        .options-grid {
            display: grid;
            grid-template-columns: 1fr 1fr;
            gap: 12px;
            margin-bottom: 20px;
        }

        .option-btn {
            background: var(--card);
            border: 2px solid #333;
            border-radius: 12px;
            padding: 16px;
            color: var(--text);
            font-size: 1.2rem;
            cursor: pointer;
            transition: all 0.3s;
            position: relative;
            overflow: hidden;
        }

        .option-btn:hover:not(.disabled) {
            border-color: var(--primary);
            background: rgba(108, 92, 231, 0.15);
            transform: translateY(-2px);
        }

        .option-btn.correct {
            border-color: var(--success);
            background: rgba(0, 184, 148, 0.2);
            animation: correctFlash 0.5s;
        }

        .option-btn.wrong {
            border-color: var(--error);
            background: rgba(214, 48, 49, 0.2);
            animation: shake 0.5s;
        }

        .option-btn.disabled {
            cursor: default;
            opacity: 0.6;
        }

        .option-btn .option-num {
            position: absolute;
            top: 6px;
            left: 10px;
            font-size: 0.65rem;
            color: var(--dim);
            font-weight: bold;
        }

        @keyframes correctFlash {
            0% { transform: scale(1); }
            50% { transform: scale(1.05); }
            100% { transform: scale(1); }
        }

        @keyframes shake {
            0%, 100% { transform: translateX(0); }
            25% { transform: translateX(-8px); }
            75% { transform: translateX(8px); }
        }

        .feedback {
            text-align: center;
            padding: 15px;
            border-radius: 12px;
            margin-bottom: 15px;
            display: none;
            font-weight: 600;
            line-height: 1.4;
        }

        .feedback.show { display: block; }
        .feedback.correct-fb {
            background: rgba(0, 184, 148, 0.15);
            border: 1px solid var(--success);
            color: var(--success);
        }
        .feedback.wrong-fb {
            background: rgba(214, 48, 49, 0.15);
            border: 1px solid var(--error);
            color: var(--error);
        }

        .feedback .correct-answer {
            font-size: 1.3rem;
            margin-top: 5px;
            display: block;
        }

        .action-btn {
            width: 100%;
            padding: 14px;
            border: none;
            border-radius: 12px;
            font-size: 1rem;
            font-weight: 600;
            cursor: pointer;
            transition: all 0.3s;
        }

        .action-btn.primary {
            background: linear-gradient(135deg, var(--accent), var(--primary));
            color: white;
        }

        .action-btn.primary:hover {
            transform: translateY(-2px);
            box-shadow: 0 8px 25px rgba(233, 69, 96, 0.3);
        }

        .action-btn.secondary {
            background: var(--card);
            border: 2px solid #333;
            color: var(--text);
            margin-top: 10px;
        }

        .action-btn.secondary:hover {
            border-color: var(--primary);
        }

        .mode-select {
            display: grid;
            grid-template-columns: 1fr 1fr;
            gap: 12px;
            margin-bottom: 20px;
        }

        .mode-btn {
            background: var(--card);
            border: 2px solid #333;
            border-radius: 16px;
            padding: 25px 15px;
            color: var(--text);
            cursor: pointer;
            transition: all 0.3s;
            text-align: center;
        }

        .mode-btn:hover {
            border-color: var(--primary);
            transform: translateY(-3px);
        }

        .mode-btn .mode-icon { font-size: 2.5rem; margin-bottom: 10px; }
        .mode-btn .mode-title { font-weight: bold; font-size: 1.1rem; margin-bottom: 5px; }
        .mode-btn .mode-desc { font-size: 0.8rem; color: var(--dim); }

        .table-container {
            overflow-x: auto;
            margin-bottom: 20px;
            border-radius: 12px;
            border: 1px solid #333;
        }

        .reaction-table {
            width: 100%;
            border-collapse: collapse;
            font-size: 0.75rem;
        }

        .reaction-table th, .reaction-table td {
            border: 1px solid #333;
            padding: 8px 4px;
            text-align: center;
            min-width: 50px;
        }

        .reaction-table th {
            background: var(--card2);
            color: var(--accent);
            font-weight: 600;
            position: sticky;
            top: 0;
        }

        .reaction-table td {
            background: var(--card);
            cursor: pointer;
            transition: all 0.2s;
        }

        .reaction-table td:hover:not(.empty):not(.filled) {
            background: rgba(108, 92, 231, 0.2);
        }

        .reaction-table td.filled {
            color: var(--success);
            font-weight: bold;
            background: rgba(0, 184, 148, 0.1);
            cursor: default;
        }

        .reaction-table td.empty {
            color: #444;
            cursor: default;
        }

        .progress-bar {
            width: 100%;
            height: 6px;
            background: #333;
            border-radius: 3px;
            margin-bottom: 20px;
            overflow: hidden;
        }

        .progress-fill {
            height: 100%;
            background: linear-gradient(90deg, var(--accent), var(--success));
            border-radius: 3px;
            transition: width 0.5s ease;
        }

        .end-screen {
            text-align: center;
            padding: 30px;
        }

        .end-screen .trophy { font-size: 4rem; margin-bottom: 15px; }
        .end-screen h2 { margin-bottom: 10px; }
        .end-screen .final-score {
            font-size: 3rem;
            font-weight: bold;
            background: linear-gradient(135deg, var(--accent), var(--primary));
            -webkit-background-clip: text;
            -webkit-text-fill-color: transparent;
        }

        .end-screen .grade {
            font-size: 1.2rem;
            color: var(--dim);
            margin: 10px 0 20px;
            white-space: pre-line;
        }

        ::-webkit-scrollbar { width: 6px; height: 6px; }
        ::-webkit-scrollbar-track { background: var(--bg); }
        ::-webkit-scrollbar-thumb { background: #333; border-radius: 3px; }

        @media (max-width: 600px) {
            .reaction-display { font-size: 1.4rem; }
            .element-box { font-size: 1.2rem; padding: 8px 12px; min-width: 60px; }
            .options-grid { grid-template-columns: 1fr; }
            .stats-bar { gap: 5px; }
            .stat-box { padding: 8px 10px; min-width: 70px; }
            .stat-box .value { font-size: 1.2rem; }
            .mode-select { grid-template-columns: 1fr; }
        }
    </style>
</head>
<body>

<div class="game-wrapper">
    <div class="header">
        <h1>🧪 Продукты реакций простых веществ</h1>
        <div style="font-size:0.75rem; color:var(--dim);">Тренажёр для запоминания реакций</div>
    </div>

    <div id="menu-screen">
        <div class="mode-select">
            <div class="mode-btn" onclick="startQuiz('random')">
                <div class="mode-icon">🎯</div>
                <div class="mode-title">Викторина</div>
                <div class="mode-desc">Выбери правильный продукт из 4 вариантов</div>
            </div>
            <div class="mode-btn" onclick="startQuiz('table')">
                <div class="mode-icon">📊</div>
                <div class="mode-title">Заполни таблицу</div>
                <div class="mode-desc">Кликай на ячейки и вписывай ответы</div>
            </div>
        </div>

        <details style="margin-bottom:20px;">
            <summary style="cursor:pointer; color:var(--accent); font-weight:600; padding:12px; background:var(--card); border-radius:10px; list-style: none;">
                📖 Справочная таблица (нажми чтобы открыть)
            </summary>
            <div class="table-container" id="ref-table-container" style="margin-top: 10px;"></div>
        </details>
    </div>

    <div id="quiz-screen" style="display:none;">
        <div class="stats-bar">
            <div class="stat-box">
                <div class="label">Верно</div>
                <div class="value green" id="correct-count">0</div>
            </div>
            <div class="stat-box">
                <div class="label">Ошибки</div>
                <div class="value" id="wrong-count">0</div>
            </div>
            <div class="stat-box">
                <div class="label">Серия</div>
                <div class="value" id="streak-count">0</div>
            </div>
            <div class="stat-box">
                <div class="label">Осталось</div>
                <div class="value" id="remaining-count">0</div>
            </div>
        </div>

        <div class="progress-bar">
            <div class="progress-fill" id="progress-fill" style="width:0%"></div>
        </div>

        <div class="reaction-card">
            <div style="font-size:0.8rem; color:var(--dim); margin-bottom: 5px;">Какой продукт образуется?</div>
            <div class="reaction-display">
                <div class="element-box" id="elem1">O₂</div>
                <span class="plus-sign">+</span>
                <div class="element-box" id="elem2">S</div>
                <span class="arrow">→</span>
                <div class="question-mark">?</div>
            </div>
            <div class="conditions-hint" id="conditions-hint"></div>
        </div>

        <div class="feedback" id="feedback"></div>
        <div class="options-grid" id="options-grid"></div>

        <button class="action-btn primary" id="submit-btn" onclick="submitAnswer()" style="display:none;">Проверить</button>
        <button class="action-btn secondary" id="next-btn" onclick="nextQuestion()" style="display:none;">Следующий вопрос →</button>
        <button class="action-btn secondary" onclick="backToMenu()" style="margin-top:10px;">← В меню</button>
    </div>

    <div id="table-screen" style="display:none;">
        <div class="stats-bar">
            <div class="stat-box">
                <div class="label">Заполнено</div>
                <div class="value green" id="table-filled">0</div>
            </div>
            <div class="stat-box">
                <div class="label">Осталось</div>
                <div class="value" id="table-remaining">0</div>
            </div>
        </div>
        <div class="progress-bar">
            <div class="progress-fill" id="table-progress" style="width:0%"></div>
        </div>
        <div class="table-container" id="game-table-container"></div>
        <button class="action-btn secondary" onclick="backToMenu()">← В меню</button>
    </div>

    <div id="end-screen" style="display:none;">
        <div class="end-screen">
            <div class="trophy" id="trophy">🏆</div>
            <h2>Результат</h2>
            <div class="final-score" id="final-score">0%</div>
            <div class="grade" id="grade-text"></div>
            <button class="action-btn primary" onclick="backToMenu()">В главное меню</button>
        </div>
    </div>
</div>

<script>
const elements = ['O₂', 'S', 'H₂', 'F₂', 'Cl₂', 'Br₂', 'I₂', 'P', 'C', 'N₂', 'Si'];

const reactions = {
    'O₂+S': { product: 'SO₂', note: '' },
    'O₂+H₂': { product: 'H₂O', note: '' },
    'O₂+F₂': { product: 'OF₂', note: '' },
    'O₂+P': { product: 'P₂O₃ / P₂O₅', note: 'P₂O₅ при избытке O₂' },
    'O₂+C': { product: 'CO / CO₂', note: 'CO₂ при избытке O₂' },
    'O₂+N₂': { product: 'NO', note: 'При высокой t°' },
    'O₂+Si': { product: 'SiO₂', note: '' },
    'S+H₂': { product: 'H₂S', note: '' },
    'S+F₂': { product: 'SF₆', note: '' },
    'S+Cl₂': { product: 'SCl₂ / S₂Cl₂', note: '' },
    'S+Br₂': { product: 'SBr₂', note: '' },
    'S+P': { product: 'P₂S₅ / P₂S₃', note: '' },
    'S+C': { product: 'CS₂', note: '' },
    'S+Si': { product: 'SiS₂', note: '' },
    'H₂+F₂': { product: 'HF', note: '' },
    'H₂+Cl₂': { product: 'HCl', note: '' },
    'H₂+Br₂': { product: 'HBr', note: '' },
    'H₂+I₂': { product: 'HI', note: 'Обратимая' },
    'H₂+C': { product: 'CH₄', note: '' },
    'H₂+N₂': { product: 'NH₃', note: 'Катализатор, t°, p' },
    'F₂+P': { product: 'PF₅', note: '' },
    'F₂+C': { product: 'CF₄', note: '' },
    'F₂+N₂': { product: 'NF₃', note: 'Особые условия' },
    'F₂+Si': { product: 'SiF₄', note: '' },
    'Cl₂+P': { product: 'PCl₃ / PCl₅', note: 'PCl₅ при избытке Cl₂' },
    'Cl₂+Si': { product: 'SiCl₄', note: '' },
    'Br₂+P': { product: 'PBr₃ / PBr₅', note: '' },
    'Br₂+Si': { product: 'SiBr₄', note: '' },
    'I₂+P': { product: 'PI₃', note: '' },
    'I₂+Si': { product: 'SiI₄', note: '' },
    'C+Si': { product: 'SiC', note: 'Карборунд' },
};

const reactionLookup = {};
for (const key in reactions) {
    reactionLookup[key] = reactions[key];
    const parts = key.split('+');
    if (parts.length === 2) {
        reactionLookup[parts[1] + '+' + parts[0]] = reactions[key];
    }
}

function getReaction(a, b) {
    return reactionLookup[a + '+' + b];
}

let currentMode = '';
let questions = [];
let currentIdx = 0;
let correctCount = 0;
let wrongCount = 0;
let streak = 0;
let maxStreak = 0;
let answered = false;

function getAllPairs() {
    const pairs = [];
    for (let i = 0; i < elements.length; i++) {
        for (let j = i + 1; j < elements.length; j++) {
            const key = elements[i] + '+' + elements[j];
            if (reactionLookup[key]) {
                // ИСПРАВЛЕНО: добавлено ""
                pairs.push({ a: elements[i], b: elements[j], data: reactionLookup[key] });
            }
        }
    }
    return pairs;
}

const allPairs = getAllPairs();

function shuffle(arr) {
    const newArr = [...arr];
    for (let i = newArr.length - 1; i > 0; i--) {
        const j = Math.floor(Math.random() * (i + 1));
        [newArr[i], newArr[j]] = [newArr[j], newArr[i]];
    }
    return newArr;
}

function showScreen(id) {
    ['menu-screen', 'quiz-screen', 'table-screen', 'end-screen'].forEach(s => {
        document.getElementById(s).style.display = 'none';
    });
    document.getElementById(id).style.display = 'block';
}

function backToMenu() {
    showScreen('menu-screen');
}

function buildRefTable() {
    const container = document.getElementById('ref-table-container');
    let html = '<table class="reaction-table"><tr><th></th>';
    elements.forEach(el => html += `<th>${el}</th>`);
    html += '</tr>';
    elements.forEach(el1 => {
        html += `<tr><th>${el1}</th>`;
        elements.forEach(el2 => {
            const rxn = getReaction(el1, el2);
            if (el1 === el2) {
                html += '<td class="empty">—</td>';
            } else if (rxn) {
                html += `<td class="filled">${rxn.product}</td>`;
            } else {
                html += '<td class="empty">·</td>';
            }
        });
        html += '</tr>';
    });
    html += '</table>';
    container.innerHTML = html;
}

function startQuiz(mode) {
    currentMode = mode;
    questions = shuffle([...allPairs]);
    currentIdx = 0;
    correctCount = 0;
    wrongCount = 0;
    streak = 0;
    maxStreak = 0;
    answered = false;
    updateStats();

    if (mode === 'table') {
        showScreen('table-screen');
        buildTableGame();
    } else {
        showScreen('quiz-screen');
        showQuestion();
    }
}

function updateStats() {
    document.getElementById('correct-count').textContent = correctCount;
    document.getElementById('wrong-count').textContent = wrongCount;
    document.getElementById('streak-count').textContent = streak;
    document.getElementById('remaining-count').textContent = questions.length - currentIdx;
    const pct = questions.length > 0 ? (currentIdx / questions.length * 100) : 0;
    document.getElementById('progress-fill').style.width = pct + '%';
}

function showQuestion() {
    if (currentIdx >= questions.length) { endGame(); return; }
    answered = false;
    const q = questions[currentIdx];

    document.getElementById('elem1').textContent = q.a;
    document.getElementById('elem2').textContent = q.b;
    document.getElementById('conditions-hint').textContent = q.data.note || '';

    const feedback = document.getElementById('feedback');
    feedback.className = 'feedback';
    feedback.style.display = 'none';

    document.getElementById('submit-btn').style.display = 'none';
    document.getElementById('next-btn').style.display = 'none';
    document.getElementById('options-grid').style.display = 'grid';

    generateOptions(q);
    updateStats();
}

function generateOptions(q) {
    const grid = document.getElementById('options-grid');
    const correct = q.data.product.split('/')[0].trim();

    const allProducts = [...new Set(allPairs.map(p => p.data.product.split('/')[0].trim()))];
    let wrongOptions = allProducts.filter(p => p !== correct);
    shuffle(wrongOptions);
    wrongOptions = wrongOptions.slice(0, 3);

    let options = [...wrongOptions, correct];
    options = shuffle(options);

    grid.innerHTML = '';
    options.forEach((opt, i) => {
        const btn = document.createElement('button');
        btn.className = 'option-btn';
        btn.innerHTML = `<span class="option-num">${i + 1}</span>${opt}`;
        btn.onclick = () => selectOption(btn, opt, correct, q);
        grid.appendChild(btn);
    });
}

function selectOption(btn, selected, correct, q) {
    if (answered) return;
    answered = true;

    const buttons = document.querySelectorAll('.option-btn');
    buttons.forEach(b => b.classList.add('disabled'));

    const feedback = document.getElementById('feedback');

    if (selected === correct) {
        btn.classList.add('correct');
        correctCount++;
        streak++;
        if (streak > maxStreak) maxStreak = streak;
        feedback.className = 'feedback correct-fb show';
        feedback.innerHTML = `✅ Верно!<br><span class="correct-answer">${q.data.product}</span>`;
    } else {
        btn.classList.add('wrong');
        buttons.forEach(b => {
            if (b.textContent.includes(correct)) b.classList.add('correct');
        });
        wrongCount++;
        streak = 0;
        feedback.className = 'feedback wrong-fb show';
        feedback.innerHTML = `❌ Неверно.<br><span class="correct-answer">Правильно: ${q.data.product}</span>`;
        if (q.data.note) feedback.innerHTML += `<br><small style="color:var(--dim)">${q.data.note}</small>`;
    }

    document.getElementById('next-btn').style.display = 'block';
    updateStats();
}

function submitAnswer() { }

function nextQuestion() {
    currentIdx++;
    showQuestion();
}

function buildTableGame() {
    const container = document.getElementById('game-table-container');
    const totalCells = allPairs.length;

    let html = '<table class="reaction-table"><tr><th></th>';
    elements.forEach(el => html += `<th>${el}</th>`);
    html += '</tr>';

    elements.forEach(el1 => {
        html += `<tr><th>${el1}</th>`;
        elements.forEach(el2 => {
            if (el1 === el2) {
                html += '<td class="empty">—</td>';
            } else {
                const rxn = getReaction(el1, el2);
                if (rxn) {
                    const cellId = el1 + '_' + el2;
                    html += `<td class="empty" id="cell_${cellId}" onclick="clickTableCell('${el1}','${el2}')">?</td>`;
                } else {
                    html += '<td class="empty" style="color:#222">·</td>';
                }
            }
        });
        html += '</tr>';
    });
    html += '</table>';
    container.innerHTML = html;

    document.getElementById('table-filled').textContent = '0';
    document.getElementById('table-remaining').textContent = totalCells;
    document.getElementById('table-progress').style.width = '0%';
}

function clickTableCell(el1, el2) {
    const rxn = getReaction(el1, el2);
    if (!rxn) return;

    const answer = prompt(`Какой продукт?\n${el1} + ${el2} → ?`);
    if (!answer) return;

    const cell = document.getElementById('cell_' + el1 + '_' + el2);
    const correct = rxn.product;

    const normalize = s => s.toLowerCase().replace(/\s/g, '').replace(/₂/g, '2').replace(/₃/g, '3').replace(/₄/g, '4').replace(/₅/g, '5');
    const correctParts = correct.split('/').map(p => normalize(p.trim()));
    const normUser = normalize(answer.trim());
    // Changed comparison to strict equality after normalization
    const isCorrect = correctParts.some(p => normUser === p);

    if (isCorrect) {
        cell.textContent = correct;
        cell.className = 'filled';
        cell.onclick = null;
        correctCount++;
        streak++;
        if (streak > maxStreak) maxStreak = streak;
    } else {
        alert(`Неверно! Правильно: ${correct}\n${rxn.note || ''}`);
        wrongCount++;
        streak = 0;
    }

    const filledCount = document.querySelectorAll('#game-table-container .filled').length;
    const totalCells = allPairs.length;
    document.getElementById('table-filled').textContent = filledCount;
    document.getElementById('table-remaining').textContent = totalCells - filledCount;
    document.getElementById('table-progress').style.width = (filledCount / totalCells * 100) + '%';
    document.getElementById('correct-count').textContent = correctCount;
    document.getElementById('wrong-count').textContent = wrongCount;
    document.getElementById('streak-count').textContent = streak;

    if (filledCount === totalCells) setTimeout(() => endGame(), 500);
}

function endGame() {
    showScreen('end-screen');
    const total = correctCount + wrongCount;
    const pct = total > 0 ? Math.round(correctCount / total * 100) : 0;
    document.getElementById('final-score').textContent = pct + '%';

    let grade = '', trophy = '';
    if (pct >= 90) { grade = '🌟 Отлично! Ты настоящий химик!'; trophy = '🏆'; }
    else if (pct >= 70) { grade = '👍 Хороший результат!'; trophy = '🥈'; }
    else if (pct >= 50) { grade = '📚 Неплохо, но нужно повторить'; trophy = '🥉'; }
    else { grade = '💪 Нужно подучить! Попробуй ещё раз'; trophy = '📖'; }

    document.getElementById('trophy').textContent = trophy;
    document.getElementById('grade-text').textContent = `${grade}\n\nПравильно: ${correctCount} | Неправильно: ${wrongCount} | Серия: ${maxStreak}`;
}

document.addEventListener('keydown', (e) => {
    if (currentMode === 'random' && !answered && e.key >= '1' && e.key <= '4') {
        const buttons = document.querySelectorAll('.option-btn');
        const idx = parseInt(e.key) - 1;
        if (buttons[idx] && !buttons[idx].classList.contains('disabled')) {
            buttons[idx].click();
        }
    }
    if (e.key === 'Enter' && answered && currentMode === 'random') {
        nextQuestion();
    }
});

buildRefTable();
</script>

</body>
</html>
'''

display(HTML(html_content))
```
